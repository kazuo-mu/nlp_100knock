## 05. n-gram

```python
sent = 'I am an NLPer'
```

```python
def ngram(sent, n=1, gram_type='word', remove= True):
    grams = []
    
    if remove:
        sent = sent.replace('.', '');
        sent = sent.replace(',', '');
        
    if (gram_type == 'word'):
        sent = sent.split(' ')
    elif (gram_type == 'str'):
        sent = sent.replace(' ', '')
    else:
        raise ValueError('gram_type is invalid')
        
    limit = len(sent)
    
    for i, word in enumerate(sent):
        if limit - i >= n: 
            gram =  sent[i:i+n]
            grams.append(gram)
        
    return grams
```

## 26. 強調マークアップの除去

```python
def rm_emphasis(string):
    
    pattern = re.compile(r'''
                (\'{2,5})
                (.*?)
                (\1)
                                        ''', re.VERBOSE + re.MULTILINE)
    
    removed  = pattern.sub(r'\2', string)
    
    return removed
```

```python
result_morphs = []

for sentence in sentences:
    morphs = []
    for chunk in sentence:
        for tok in chunk:
            surface = tok.text
            
            feature = tok.attrib['feature'].split(',')
            pos = feature[0]
            pos1 = feature[1]
            base = feature[6]
            
            morph = Morph(surface, base, pos, pos1)
            morphs.append(morph)
            
    result_morphs.append(morphs)
            
```

## 41. 係り受け解析結果の読み込み（文節・係り受け）

```python
class Chunk:
    def __init__(self, cid, morphs, dst, srcs):
        self.id = cid
        self.morphs = morphs
        self.dst = dst
        self.srcs = srcs
```

```python
sentences_of_chunks = []

for sentence in sentences:
    chunks = []
    chunk_info = {}
    
    # チャンクに関する情報をいったん辞書へ
    for chunk in sentence:
        
        cid = chunk.attrib['id']
        link = chunk.attrib['link']
        
        # 初期化
        if cid not in chunk_info.keys():
            chunk_info[cid] = { 'dst':  '', 'srcs': [], 'morphs': [] }
        if (link != '-1') and (link not in chunk_info.keys()):
            chunk_info[link] = { 'dst':  '', 'srcs': [], 'morphs': [] }
        
        # 辞書へ登録
        chunk_info[cid]['dst'] = link
        
        if link != '-1': 
            chunk_info[link]['srcs'].append(cid)
            
        morphs = []
        for tok in chunk:
            surface = tok.text
            feature = tok.attrib['feature'].split(',')
            pos = feature[0]
            pos1 = feature[1]
            base = feature[6]
            morph = Morph(surface, base, pos, pos1)
            morphs.append(morph)
        chunk_info[cid]['morphs'] = morphs

    # 辞書からオブジェクトを生成
    for chunk_id in chunk_info.keys():
        chunk = Chunk(chunk_id, chunk_info[chunk_id]['morphs'], chunk_info[chunk_id]['dst'], chunk_info[chunk_id]['srcs'])
        chunks.append(chunk)
        
    sentences_of_chunks.append(chunks)
```

## 42. 係り元と係り先の文節の表示


```python
depends42 = []

for sentence in sentences_of_chunks:
    
    # 係り元
    for chunk in sentence:
        frm = ''.join([morph.surface for morph in chunk.morphs]).replace('。', '').replace('、', '')
        dst = chunk.dst
        if dst == '-1':
            to = 'x'
            continue
        
        ## 係り先
        for chunk in sentence:
            if chunk.id == dst:
                to = ''.join([morph.surface for morph in chunk.morphs]).replace('。', '').replace('、', '')
                break
        
        depends42.append('{}\t{}'.format(frm, to))
```

## 43. 名詞を含む文節が動詞を含む文節に係るものを抽出

```python
depends43 = []

for sentence in sentences_of_chunks:
    
    # 係り元
    for chunk in sentence:
        
        pos_src = [morph.pos for morph in chunk.morphs]
        dst = chunk.dst
        
        ## 名詞を含み、かつ、係り先があれば
        if ('名詞' in pos_src) and (dst != '-1'):
            frm = ''.join([morph.surface for morph in chunk.morphs]).replace('。', '').replace('、', '')
        else:
            continue
            
        ## 係り先
        for chunk in sentence:
            if chunk.id == dst:
                pos_dst = [morph.pos for morph in chunk.morphs]
                
                if '動詞' in pos_dst:
                    to = ''.join([morph.surface for morph in chunk.morphs]).replace('。', '').replace('、', '')
                    break
                else:
                    continue
            else:
                continue
        
        depends43.append('{}\t{}'.format(frm, to))
```

## 45. 動詞の格パターンの抽出

```python
verb_cases_list = []

for sentence in sentences_of_chunks:
    
    # 係り先
    for chunk in sentence:
        
        # 動詞が含まれているか
        verb = [morph.base for morph in chunk.morphs if morph.pos == '動詞']
        
        if verb: 
            one_verb = verb[0]
        else:
            continue
            
        # 係り元が存在するか
        srcs = chunk.srcs
        if not srcs: continue
        
        # 係元から助詞を取得
        particles = []
        for chunk_src in sentence:
            
            # 係元について
            if chunk_src.id in srcs:
                particle = [morph.base for morph in chunk_src.morphs if morph.pos == '助詞']
                
                # 助詞があれば先頭だけ
                if particle:
                    one_particle = particle[0]
                else:
                    continue
                    
                particles.append(one_particle)
            
        if particles: 
            particles_str = ' '.join(particles)
        else:
            continue
            
        verb_cases_list.append('{}\t{}'.format(one_verb, particles_str))
```

## 46. 動詞の格フレーム情報の抽出

```python
verb_cases_list = []

for sentence in sentences_of_chunks:
    
    # 係り先
    for chunk_dst in sentence:
        
        # 動詞が含まれているか
        verb = [morph.base for morph in chunk_dst.morphs if morph.pos == '動詞']
        
        if verb: 
            result_verb = verb[0]
        else:
            continue
            
        # 係り元が存在するか
        srcs = chunk_dst.srcs
        if not srcs: continue
        
        # 係元から助詞を取得
        particles = []
        chunks = []
        
        for chunk_src in sentence:
            
            # 係元について
            if chunk_src.id in srcs:
                particle = [morph.base for morph in chunk_src.morphs if morph.pos == '助詞']
                
                # 助詞があれば、文節を取得
                if particle:
                    one_particle = particle[0]
                    one_chunk = ''.join([morph.surface for morph in chunk_src.morphs])
                    
                    particles.append(one_particle)
                    chunks.append(one_chunk)
                
                else:
                    continue
                    
        if particles and chunks: 
            result_particles = ' '.join(particles)
            result_chunks = ' '.join(chunks).replace('。', '').replace('、', '')
        else:
            continue
            
        verb_cases_list.append('{}\t{}\t{}'.format(result_verb, result_particles, result_chunks))
```

## 47. 機能動詞構文のマイニング (未完)

```python
verb_cases_list47 = []

for sentence in sentences_of_chunks:
    
    # 係り先
    for chunk_dst in sentence:
        
        # 動詞が含まれているか
        verb = [morph.base for morph in chunk_dst.morphs if morph.pos == '動詞']
        
        if verb:
            result_verb = verb[0]
        else:
            continue
            
        # 係り元があれば
        srcs = chunk_dst.srcs
        if not srcs:
            continue
        
        # 係り元から助詞を取得
        particles = []
        chunks = []
        
        for chunk_src in sentence:
            
            # 係り元について
            if chunk_src.id in srcs:
                
                # サ変接続名詞が含まれているか
                sahen = [morph.surface for morph in chunk_dst.morphs if morph.pos == '名詞' and morph.pos1 == 'サ変接続' ]
        
                # 助詞「を」が含まれているか
                wo = [morph.surface for morph in chunk_dst.morphs if morph.surface == 'を' and morph.pos == '助詞']
                
                if sahen and wo:
                    result_sahen = sahen[0]
                    result_wo = wo[0]
                else:
                    continue
                    
                particle = [morph.base for morph in chunk_src.morphs if morph.pos == '助詞']
                
                # 助詞があれば、文節を取得
                if particle:
                    one_particle = particle[0]
                    one_chunk = ''.join([morph.surface for morph in chunk_src.morphs])
                    
                    particles.append(one_particle)
                    chunks.append(one_chunk)
                
                else:
                    continue
                    
        if particles and chunks: 
            result_particles = ' '.join(particles)
            result_chunks = ' '.join(chunks).replace('。', '').replace('、', '')
        else:
            result_particles = 'x'
            result_chunks = 'x'
            
        result_predicate = result_sahen + result_wo + result_verb
            
        verb_cases_list47.append('{}\t{}\t{}'.format(result_predicate, result_particles, result_chunks))
```
